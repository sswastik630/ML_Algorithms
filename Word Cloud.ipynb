```python
import wikipedia
import re
```

```python
wiki=wikipedia.page('Shah Rukh khan')    #Specify the title of the wikipedia page
```

```python
text=wiki.content            #Extract content of wikipedia page as plain text
```

```python
text= re.sub(r'==,*?==+','',text)                 #Clean text 
```

```python
text=text.replace('\n','')
```

```python
#Visualization
import matplotlib.pyplot as plt
 #def a function to plot word cloud
def cloud(wordCloud):
    plt.figure(figsize=(40,30))
    plt.imshow(wordCloud)
    plt.axis("off")
    plt.show()
```

```python
#Lets create word cloud and then plot it.
from wordcloud import WordCloud,STOPWORDS
```

```python
!pip install wordcloud
```

```python
wordcloud= WordCloud(width = 3000, height=2000, random_state=5, background_color='salmon', colormap='Pastel1',collocations=False
                    ,stopwords=STOPWORDS).generate(text)
```

```python
cloud(wordcloud)
```
