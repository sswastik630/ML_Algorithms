Create subplot 

Matplotlib allows us easily create multi-plots on the same figure using the .subplot() method.This .subplot() method takes in three parameters, namely:
nrows: the number of rows the Figure should have.
ncols: the number of columns the Figure should have.
plot_number : which refers to a specific plot in the Figure.

```python
# .sublot(nrow, ncol,plotno)
plt.subplot(1,2,1)
plt.plot(x,y,'red')

plt.subplot(1,2,2)
plt.plot(y,x,'green')
```

##Object oriented Interface:
 This is the best way to create plots. The idea here is to create Figure objects and call methods off it. Letâ€™s create a blank Figure using the .figure() method

```python
fig=plt.figure()
```

Now we need to add a set of axes to it using the .add_axes() method.

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.5,0.8,0.9])
```

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.5,0.8,0.9])
ax.plot(x,y,'purple')
```

We can further add x and y labels and a title to our plot same way we did in the Function approach, but thereâ€™s a slight difference here. Using .set_xlabel(), .set_ylabel() and .set_title()

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.5,0.8,0.9])
ax.set_xlabel('X label')
ax.set_ylabel('Y_label')
ax.set_title('Second Plot')
ax.plot(x,y,'purple')
```

 Letâ€™s try to put in two sets of figures on one canvas:

Like we did in the functional approach, we can also create multiple plots in the object-oriented approach using the .subplots() method, and NOT .subplot(). The .subplots() method takes in nrows, which is the number of rows the Figure should have, and ncols, the number of columns the Figure should have

```python
fig, axes=plt.subplots(3,3)
```

```python
fig, axes=plt.subplots(3,3)
plt.tight_layout()    #, there is an issue of overlapping in the subplots we created. We can deal with that by using .tight_layout() method to space it out:
```

The only difference between plt.figure() and plt.subplots() is that plt.subplots() automatically does what the .add_axes() method of .figure() will do for you based off the number of rows and columns you specify

```python
fig, axes=plt.subplots(3,3)
plt.tight_layout()  

axes[0,1].plot(x,y)
axes[1,2].plot(y,x)
plt.tight_layout()
```

```python
fig, axes=plt.subplots(3,3)
plt.tight_layout()  

axes[0,1].plot(x,y)
axes[0,1].set_title('Increasing')
axes[1,2].plot(y,x)
axes[1,2].set_title('Decreasing')
plt.tight_layout()
```

### Figure size, aspect ratio, and DPI
Matplotlib allows us create customized plots by specifying the figure size, aspect ratio, and DPI by simply specifying the figsize and dpi arguments. The figsize is a tuple of the width and height of the figure (in inches), and dpi is the dots-per-inch (pixel-per-inch).

```python
fig=plt.figure(figsize=(8,2), dpi=100)
ax=fig.add_axes([0.1,0.5,0.8,0.9])
ax.plot(x,y,'purple')
```

```python
fig=plt.figure(figsize=(8,2), dpi=1000)
ax=fig.add_axes([0.1,0.5,0.8,0.9])
ax.plot(x,y,'purple')
```

## Legends

 to identify the plots, we need to add a legend using .legend() and then specify the label=â€ â€ attribute for each plot

```python
fig=plt.figure(figsize=(8,2), dpi=100)
ax=fig.add_axes([0.1,0.1,0.8,0.9])
ax.plot(x,x**2,'purple',label='square')
ax.plot(x,x**3,'red',label='cube')
plt.legend()
```

Matplotlib gives us a lot of options for customizing the appearance of our plots. By now, you should be familiar with changing line color using color=â€™redâ€™ or â€˜redâ€™ like we did in previous examples. Now we want to change linewidth or lw , linestyle or ls, and mark out data points using marker.
[link text](https://matplotlib.org/stable/api/markers_api.html)

```python
fig=plt.figure(figsize=(8,2), dpi=100)
ax=fig.add_axes([0.1,0.1,0.8,0.9])
ax.plot(x,x**2,'purple',linewidth=3,linestyle='--',marker='o',label='square')
plt.legend()
```

```python
#.set_xlim()
#.set_ylim()
fig=plt.figure(figsize=(8,2), dpi=100)
ax=fig.add_axes([0.1,0.1,0.8,0.9])
ax.set_xlim(0,5)
ax.set_ylim(0,50)
ax.plot(x,x**2,'purple',linewidth=3,linestyle='--',marker='o',label='square')
plt.legend()
```

##Scatter plots
Scatter plots offer a convenient way to visualize how two numeric values are related in your data. It helps in understanding relationships between multiple variables. Using .scatter() method, we can create a scatter plot:

```python

fig=plt.figure(figsize=(8,2), dpi=100)
ax=fig.add_axes([0.1,0.1,0.8,0.9])
ax.set_xlim(2,10)
ax.set_ylim(0,100)
ax.scatter(x,x*9)
```
